```python
def process_data(data):

    lines = [line for line in data.split("\n") if line]

    names = lines[::3]
    tickers = lines[1::3]
    inception_dates = [dt.datetime.strptime(date, '%m/%d/%Y').date() for date in lines[2::3]]
    end_date = dt.date(2023, 7, 31)

    dataframe = pd.DataFrame({
        'Name': names,
        'Ticker': tickers,
        'Inception Date': inception_dates,
        'End Date': [end_date for _ in inception_dates]
    })

    return dataframe
```

```python
def get_fund_results(fund, dataframe):

    df = fmp.get_historical_price(fund)
    df = df.iloc[::-1].reset_index(drop=True)
    df['date'] = pd.to_datetime(df['date']) 
    df.set_index('date', inplace=True)
    
    # Set the start date based on the earliest date available in the data
    earliest_data_date = df.index.min().date()
    start_date_threshold = dt.date(2018, 10, 1)
    if earliest_data_date > start_date_threshold:
        start = earliest_data_date
    else:
        start = start_date_threshold

    # Set the end date to 2023-07-31    
    end = dt.date(2023, 7, 31)

    df = df[start:end]
    daily = df[['adjClose']].pct_change()
    monthly = daily.resample('M').agg(lambda x: (x+1).prod() - 1)
    monthly.rename(columns={'adjClose': fund}, inplace=True)
    monthly = monthly.iloc[1:]

    # Slice the factors dataset based on the start and end dates
    start_period = pd.Period(start, freq='M')
    end_period = pd.Period(end, freq='M')
    factors = factors_all[start_period:end_period].iloc[1:]

    monthly.index = factors.index
    merged = pd.merge(monthly, factors, on='Date')
    merged[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']] = merged[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']]/100
    merged[f'{fund}-RF'] = merged[fund] - merged['RF']

    y = merged[f'{fund}-RF']
    X = merged[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']]
    X_sm = sm.add_constant(X)
    model = sm.OLS(y,X_sm)
    results = model.fit()

    # Extract the desired results
    description = dataframe[dataframe['Ticker'] == fund]['Name'].iloc[0]
    inception_date = dataframe[dataframe['Ticker'] == fund]['Inception Date'].iloc[0].strftime('%Y-%m-%d')
    alpha = results.params['const']
    mkt_rf = results.params['Mkt-RF']
    smb = results.params['SMB']
    hml = results.params['HML']
    rmw = results.params['RMW']
    cma = results.params['CMA']
    r_squared = results.rsquared

    return {
        'Ticker': fund,
        'Description': description,
        'Inception Date': inception_date,
        'Alpha': alpha,
        'Mkt-RF': mkt_rf,
        'SMB': smb,
        'HML': hml,
        'RMW': rmw,
        'CMA': cma,
        'R-squared': r_squared
    }
```

```python
ticker_df = process_data(data)
ticker2_df = process_data(data2)

datasets = get_available_datasets()
df_5_factor = [dataset for dataset in datasets if '5' in dataset and 'Factor' in dataset]
ds_factors = web.DataReader(df_5_factor[0], 'famafrench')
factors_all = ds_factors[0]

tickers = ticker_df['Ticker'].tolist()

results_list = []
for fund in tickers:
    try:
        results_list.append(get_fund_results(fund, ticker_df))
    except Exception as e:
        print(f"Failed to get results for {fund}. Error: {e}")

results_df = pd.DataFrame(results_list)

tickers2 = ticker2_df['Ticker'].tolist()
results2_list = []
for fund in tickers2:
    try:
        results2_list.append(get_fund_results(fund, ticker2_df))
    except Exception as e:
        print(f"Failed to get results for {fund}. Error: {e}")

results2_df = pd.DataFrame(results2_list)

results_df['Type'] = 'ETF'
results2_df['Type'] = 'Mutual Fund'


columns = results_df.columns.tolist()
columns = columns[:2] + [columns[-1]] + columns[2:-1]
results_df = results_df[columns]

columns = results2_df.columns.tolist()
columns = columns[:2] + [columns[-1]] + columns[2:-1]
results2_df = results2_df[columns]
```
